4.MOBILE RECHARGE VALIDITY

```python

from datetime import datetime, timedelta
def check_plan_validity(recharge_date_str, today_date_str):
    try:
        #  the input dates (format: YYYY-MM-DD)
        recharge_date = datetime.strptime(recharge_date_str, "%Y-%m-%d").date()
        today_date = datetime.strptime(today_date_str, "%Y-%m-%d").date()

        # Validate that today is not before recharge date
        if today_date < recharge_date:
            print("Error: Today's date cannot be before the recharge date.")
            return

        # Plan validity period (28 days)
        expiry_date = recharge_date + timedelta(days=28)

        # Check if plan is still valid
        if today_date <= expiry_date:
            remaining_days = (expiry_date - today_date).days
            print(f" Plan is still valid. {remaining_days} day(s) remaining. Expiry date: {expiry_date}")
        else:
            print(f" Plan has expired.Recharge again. Expired on: {expiry_date}")

    except ValueError:
        print("Invalid date format. Please enter dates in YYYY-MM-DD format.")

if __name__ == "__main__":
    # Take user input
    recharge_date_input = input("Enter recharge date (YYYY-MM-DD): ").strip()
    today_date_input = input("Enter today's date (YYYY-MM-DD): ").strip()

    check_plan_validity(recharge_date_input, today_date_input)



```

5.BUS SEAT RESERVATION SYSTEM

```python

a=[]
n = int(input("Enter the number of names wanted to enter: "))
for i in range(n):
    element = input(f"Enter booked person {i+1}: ")
    a.append(element)

print("booked:", a)
r = str(input("Enter the name to be removed : "))
if r in a :
  index_removed = a.index(r)
  a.remove(r)
  print("Cancel= ",a)
else:
  print(f"No booking found for {r}.")
n = str(input("Enter the new name  : "))
a.insert(index_removed, n)
#print(f" Booking for {n} added at position {index_removed}.")

print(" new list:", a)
```

Dictionary Word frequency

```python
sentence = input("Enter a sentence: ") # Convert to lowercase
sentence = sentence.lower() # Split into words
words = sentence.split()
 # Create empty dictionary
freq = {}
  # Count using dictionary get() method
for word in words:
 freq[word] = freq.get(word, 0) + 1
   # Display result
print("Word frequency:")
for word, count in freq.items():
 print(word, ":", count)
```

Function check prime

```python
def is_prime(n):
   # Prime numbers are greater than 1
    if n <= 1:
       return False # Check prime or not
    for i in range(2, n):
      if n % i == 0:
        return False

    return True
     # Test the function
num = int(input("Enter a number: "))
if is_prime(num):
    print(num, "is a prime number")
else:
    print(num, "is not a prime number")
```

9.Vowel counter


```python
def vc(n):
    vowels = ["a", "e", "i", "o", "u"]
    count = 0

    for ch in n:
        if ch.lower() in vowels:
            print(ch)
            count += 1

    return count


n = input("Enter a string: ")
result = vc(n)
print("Total vowels:", result)
```
